```python
import numpy as np
from math import ceil,floor
from scipy.special import erfc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy.integrate import simps

def make_bins(sample,percent=0.01):
    start,stop = sample.min(),sample.max()
    start,stop = floor(start),ceil(stop)
    bin_width = round((stop-start)*percent,3)
    bins = np.arange(start-bin_width,stop+bin_width,bin_width)
    return bins,bin_width

def integrate(bin_width,X,start=0):
    integral = np.sum(X[start:]*bin_width)
    return integral

def ROC(bin_width,X,Y):
    assert X.shape==Y.shape, "Non-matching shapes of the distributions!"
    n_bins = X.shape[0]
    roc = np.empty((n_bins,2))
    for i in xrange(n_bins):
        IX = integrate(bin_width,X,start=i)
        IY = integrate(bin_width,Y,start=i)
        roc[i] = IX,IY
    return roc

def AUROC(roc):
    x,y = roc[::-1,0],roc[::-1,1]
    widths = np.diff(x)
    heights = y[:-1]+np.diff(y)/2
    area = np.sum(heights*widths)
    if area<0.5:
        area = 1.0-area 
    return area

def gauss_AUROC(g1,g2):
    """
    Require dos argumentos, cada uno una tupla que contenga las caracetristicas de una gaussiana:
    la primera entrada de la tupla debe ser el promedio y la segunda la desviación estándar.
    Ref: Carnevale et al, 2013
    """
    D = abs(g2[0]-g1[0])/np.sqrt(0.5*(g1[1]+g2[1]))
    return 0.5*erfc(-0.5*D)
```

```python
N = 20 # tamaño de la muestra
p = 0.05 # minimo de probabilidad para ser significativo
shuffles = 1000 # numero de permutaciones
iterations = 1000 # numero de muestreos
fo = 0.8 # factor de fano
mu = [20.0,22.0] # medias de las distribuciones a comparar
```

```python
param1 = [mu[0],fo*mu[0]]
param2 = [mu[1],fo*mu[1]]
sample1 = np.random.normal(loc=param1[0],scale=np.sqrt(param1[1]),size=(N,))
sample2 = np.random.normal(loc=param2[0],scale=np.sqrt(param2[1]),size=(N,))
```

```python
bins = make_bins(np.concatenate((sample1,sample2)))
distribution1 = np.histogram(sample1,bins=bins[0],density=True)[0]
distribution2 = np.histogram(sample2,bins=bins[0],density=True)[0]
x = bins[0][0:-1]+bins[1]/2
print len(bins[0]),bins[1]
```

```python
plt.bar(left=bins[0][:-1],width=0.5,height=distribution1,color="blue",alpha=0.7);
plt.bar(left=bins[0][:-1],width=0.5,height=distribution2,color="orange",alpha=0.7);
```

```python
roc = ROC(bins[1],distribution1,distribution2)
auroc = AUROC(roc)
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111)
ax.plot(roc[:,0],roc[:,1],lw=1.5)
ax.plot(roc[:,0],roc[:,1],".",color="blue",alpha=0.5)
ax.plot(np.linspace(0,1),np.linspace(0,1),"--",color="black")
ax.set_xlim(-0.1,1.0),ax.set_ylim(0.0,1.1),ax.grid(True);
# these are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
# place a text box in upper left in axes coords
textstr = "$AUROC = {}$".format(str(auroc))
plt.text(0.65, 0.25, textstr, transform=ax.transAxes, fontsize=14, verticalalignment='top', bbox=props);
```

## Permutation Test

```python
p = 0.0

all_samples = np.concatenate((sample1,sample2))
indices = range(0,2*N)
auroc_sample = np.empty((iterations,))
for i in xrange(iterations):
    idx1 = np.random.choice(indices,size=(N,),replace=False)
    idx2 = np.array(list(set(indices).difference(set(idx1))))
    shuffle1,shuffle2 = all_samples[idx1], all_samples[idx2]
    dist1 = np.histogram(shuffle1,bins=bins[0],density=True)[0]
    dist2 = np.histogram(shuffle2,bins=bins[0],density=True)[0]
    shuffled_roc = ROC(bins[1],dist1,dist2)
    shuffled_auroc = AUROC(shuffled_roc)
    auroc_sample[i] = shuffled_auroc
    if shuffled_auroc>=auroc:
        p += 1.0
p = p/iterations        
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111)
ax.hist(auroc_sample,bins=30,alpha=0.9,color="orange")
ax.vlines(auroc,*ax.get_ylim(),lw=2.0,linestyle="--")
# these are matplotlib.patch.Patch properties
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
# place a text box in upper left in axes coords
textstr = "$p = {}$".format(str(p))
ax.text(0.80, 0.85, textstr, transform=ax.transAxes, fontsize=14, verticalalignment='top', bbox=props);
```
