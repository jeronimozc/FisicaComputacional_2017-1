Volvemos a pleantear las ecuaciones de manera análoga al Problema 1, pero agregamos el término de la fuerza de Coulomb,

$F_{c} = \frac{q^{2}}{r^{3}}\vec r$

a nuestra ecuación de movimiento.

```python
def ec_mov(r_v, t, k = 1.0, m = 1.0, q = 1.0, fric=0.0):
    """
    r_v contiene r, theta, r punto, theta punto
    ecuacion de movimiento con gravedad radial
    """
    x,y   = r_v[0:2]
    dx,dy = r_v[2:4]    
    r  = sqrt(r_v[0]**2+r_v[1]**2)
    if fric == 0.0:
        dvx = x*( q*q/r**3 -(k/m) )
        dvy = y*( q*q/r**3 -(k/m) )
    else:
        ff = fric([dx, dy])
        dvx = x*( q*q/r**3 -(k/m) ) -ff[0]
        dvy = y*( q*q/r**3 -(k/m) ) -ff[1]  
    return array([dx,dy,dvx,dvy])
def fric(v, gamma=0.1, mu=0.2):
    norma = sqrt(v[0]**2+v[1]**2)
    if norma <= 1:
        f_x = gamma*v[0]
        f_y = gamma*v[1]
    else:
        f_x = mu*norma**(1.5)*np.sign(v[0])
        f_y = mu*norma**(1.5)*np.sign(v[1])
    return array([f_x,f_y])
```

Aqui muestro un ejemplo de una órbita claramente cerrada. En prinicipio, como todas la fuerzas involucradas son conservativas y centrales, podemos tener orbitas cerradas. Pero como nuestro oscilador no tiene la forma adecuada según el teorema de Bertrand, no todas las orbitas acotadas serán cerradas.

Para el caso con fricción, podemos claramente decir que no habrá órbitas cerradas ya que la fricción introduce un término no conservativo. La partícula perderá momento hasta llegar a un punto de equilibrio, en este caso, entre la fuerza repulsiva de Coulomb, y la atracción del oscilador.
