```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
%run lib.ipynb
%matplotlib inline
```

Primero defino las funciones para la ecuacion de movimiento y el modelo de fricci√≥n dependiente de $|v|$

```python
def ec_mov(r_v, t, k = 1.0, m = 1.0, fric=0.0):
    """
    r_v contiene r, theta, r punto, theta punto
    ecuacion de movimiento con gravedad radial
    """
    x,y   = r_v[0:2]
    dx,dy = r_v[2:4]    
    r  = sqrt(r_v[0]**2+r_v[1]**2)
    if fric == 0.0:
        dvx = -(k/m)*x
        dvy = -(k/m)*y
    else:
        ff = fric([dx, dy])
        dvx = -(k/m)*x -ff[0]
        dvy = -(k/m)*y -ff[1]  
    return array([dx,dy,dvx,dvy])
def fric(v, gamma=0.1, mu=0.2):
    norma = sqrt(v[0]**2+v[1]**2)
    if norma <= 1:
        f_x = gamma*v[0]
        f_y = gamma*v[1]
    else:
        f_x = mu*norma**(1.5)*np.sign(v[0])
        f_y = mu*norma**(1.5)*np.sign(v[1])
    return array([f_x,f_y])
```

## (B) Soluciones para multiples angulos de disparo 

Los angulos de la velocidad de salida son $n\frac{\pi}{6}$ para $n\in\{0,...,11\}$, con $|v|=1$

```python
t, h = [0,20], 0.1
solutions = {}
for n in xrange(12):
    p = [0,1,np.cos((np.pi*n/6)),np.sin((np.pi*n/6))]
    sol= int_rk4(ec_mov, p, t, h, fric=fric)[1]
    solutions[n] = sol
```

## ( C ) Energia

Ahora definimos la energ√≠a como lo que tiene que ser.

```python
def energia(r_v, k = 1.0, m = 1.0):
    K = 0.5*m*r_v[2]**2+0.5*m*r_v[3]**2
    R = 0.5*k*r_v[0]**2+0.5*k*r_v[1]**2
    return K+R
```

```python
t, h = [0,20], 0.1
solutions = {}
for n in xrange(12):
    inicio = [0,1,np.cos((np.pi*n/6)),np.sin((np.pi*n/6))]
    trayectoria = [inicio]
    E0 = energia(inicio)
    energias = [E0]
    i = 1
    while energia(trayectoria[i-1])>=0.001*E0:
        t,sol = int_rk4(ec_mov,trayectoria[i-1],[0,h],fric=fric)
        y = sol[-1,:]
        trayectoria.append(y)
        energias.append(energia(y))
        i += 1
    solutions[n] = [np.array(trayectoria),np.array(energias)]
```

```python
fig = plt.figure(figsize=(14,7))

ax1 = fig.add_subplot(121)
for n in xrange(12):
    sol, enr = solutions[n]
    colorline(sol[:,0], sol[:,1],cmap='copper', linewidth=1);
    ax1.plot(sol[ 0,0],sol[ 0,1],'o-',c="b",ms=8,alpha=0.8)
    ax1.plot(sol[-1,0],sol[-1,1],'o-',c="r",ms=8,alpha=0.8)
ax2 = fig.add_subplot(122)    
# energ√≠as
for n in xrange(12):
    sol, enr = solutions[n]
    ax2.plot(range(enr.shape[0]), enr, linewidth=1, label="n={}".format(n));
    
ax2.grid(True);ax1.grid(True);
ax1.set_title("Trayectorias");ax2.set_title("Energias");
ax2.legend();
```

## (D) Momento angular, $l=r\times p$

```python
ang_momt = {}
for n in xrange(12):
    sol = solutions[n][0]
    ang_momt[n] = np.cross(sol[:,[0,1]],sol[:,[2,3]])
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
for n in xrange(12):
    sol = solutions[n][0]
    mom = np.cross(sol[:,[0,1]],sol[:,[2,3]])
    ax.plot(range(mom.shape[0]),mom,'-',ms=8,alpha=0.8, label="n = {}".format(n))
    ax.plot(range(mom.shape[0])[-1],mom[-1],'o-',c="r",ms=8,alpha=0.8)    
    ax.plot(range(mom.shape[0])[ 0],mom[ 0],'o-',c="g",ms=8,alpha=0.8)    
    
ax.grid(True); ax.set_title("Momento angular"); ax.legend(loc=(1.0,0.0));
ax.set_xlim(-10,700);ax.set_ylim(-1.1,1.1);
```

## (E) Espacios fase

```python
t, h, v = [0,50], 0.01, 4
solutions = {}
for n in xrange(4):
    a   = 2*np.pi*np.random.randn()
    x,y = np.random.randn(),np.random.randn()
    p = [x,y,v*np.cos(a),v*np.sin(a)]
    sol= int_rk4(ec_mov, p, t, h, fric=fric)[1]
    solutions[n] = sol
```

```python
fig = plt.figure(figsize=(14,7))
ax_x = fig.add_subplot(121)
ax_r = fig.add_subplot(122)
for n in xrange(4):
    sol = solutions[n]
    r  = norm(sol[:,[0,1]],axis=1)
    Pr = (sol[:,0]*sol[:,2]+sol[:,1]*sol[:,3])/r
    ax_r.plot(r,Pr,'-',ms=8,alpha=0.8, label="Ej. {}".format(n+1))
    ax_r.plot(r[-1],Pr[-1],'o-',c="r",ms=8,alpha=0.8)    
    ax_r.plot(r[ 0],Pr[ 0],'o-',c="g",ms=8,alpha=0.8)
    # espacio fase de x x_dot
    ax_x.plot(sol[:,0],sol[:,2],'-',ms=8,alpha=0.8)
    ax_x.plot(sol[-1][0],sol[-1][2],'o-',c="r",ms=8,alpha=0.8)    
    ax_x.plot(sol[ 0][0],sol[ 0][2],'o-',c="g",ms=8,alpha=0.8)
    
ax_r.grid(True); ax_r.set_title("Espacio fase radial"); ax_r.legend(loc=(1.0,0.0));
ax_r.set_ylabel("$P_{r}$",rotation=0.0);ax_r.set_xlabel("$r$");
ax_x.grid(True); ax_x.set_title("Espacio fase x");
ax_x.set_ylabel("$P_{x}$",rotation=0.0);ax_x.set_xlabel("$x$");
```
